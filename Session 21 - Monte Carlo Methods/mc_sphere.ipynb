**Monte Carlo estimation of the unit sphere volume using the Halton sequence.**
Extending the 2D circle approach to 3D: a unit sphere (radius 1) inscribed
in a $2\times 2\times 2$ cube has volumes:

$$V_{\text{sphere}} = \frac{4}{3}\pi r^3 = \frac{4}{3}\pi
\qquad
V_{\text{cube}} = (2r)^3 = 8$$

Scattering random points uniformly over the cube, the fraction inside the sphere
equals the volume ratio:

$$\frac{\text{dots inside}}{\text{dots total}} \approx \frac{4\pi/3}{8}
\quad\Rightarrow\quad
\frac{4}{3}\pi \approx 8 \times \frac{\text{dots inside}}{\text{dots total}}$$

---
**Halton sequence (Numba-compiled).**
Same algorithm as `mc_circle_halton.ipynb`;
extended here to three dimensions by applying it with a third prime base (5).

A point $(x,y,z)$ is inside the unit sphere if and only if $x^2+y^2+z^2 \leq 1$.
The Halton sequence uses three distinct prime bases (2, 3, 5) for the
three coordinates to ensure each dimension is filled uniformly and
independently.

```python
"""mc_sphere.ipynb"""

# Cell 01 - Numba-compiled Halton quasi-random sequence

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numba import float64, int64, vectorize


@vectorize([float64(int64, int64)], nopython=True)
def halton(n, p):
    h, f = 0, 1
    while n > 0:
        f = f / p
        h += (n % p) * f
        n = int(n / p)
    return h


print(halton(np.int64(10_000), np.int64(2)))
```

**Sample count.**
125,000 points - comparable to the circle notebooks but now distributed
across three dimensions.

```python
# Cell 02 - Number of sample points

total_dots = 125_000
print(f"{total_dots = :,}")
```

**Generating 3D Halton sample points.**
$x$ uses base 2, $y$ uses base 3, $z$ uses base 5 - three distinct primes
so the three coordinates are mutually independent.
Each coordinate is scaled from $(0,1]$ to $(-1, 1]$.

```python
# Cell 03 - Generate 3D Halton sample points in [-1, 1]^3
# x: base 2, y: base 3, z: base 5 (distinct primes -> independent dimensions)

n_arr = np.arange(total_dots, dtype=np.int64)

x = (1 - halton(n_arr, np.full(total_dots, 2, dtype=np.int64))) * 2 - 1
y = (1 - halton(n_arr, np.full(total_dots, 3, dtype=np.int64))) * 2 - 1
z = (1 - halton(n_arr, np.full(total_dots, 5, dtype=np.int64))) * 2 - 1

pd.DataFrame({"x": x[:5], "y": y[:5], "z": z[:5]})
```

**Distance from the origin.**
Extended to 3D: $d = \sqrt{x^2 + y^2 + z^2}$.
A point is inside the unit sphere if and only if $d \leq 1$.

```python
# Cell 04 - Euclidean distance from the origin for every 3D sample point

d = np.sqrt(x**2 + y**2 + z**2)

pd.DataFrame({"x": x[:5], "y": y[:5], "z": z[:5], "d": d[:5]})
```

**Classifying points as inside or outside the sphere.**
Same boolean indexing as the 2D notebooks, extended to the $z$ coordinate.

```python
# Cell 05 - Partition points: inside (d <= 1) vs outside (d > 1)

x_in = x[d <= 1.0]
y_in = y[d <= 1.0]
z_in = z[d <= 1.0]
x_out = x[d > 1.0]
y_out = y[d > 1.0]
z_out = z[d > 1.0]

pd.DataFrame(
    {
        "x_in": x_in[:5],
        "y_in": y_in[:5],
        "z_in": z_in[:5],
        "x_out": x_out[:5],
        "y_out": y_out[:5],
        "z_out": z_out[:5],
    }
)
```

**3D scatter plot.**
Red points are inside the unit sphere, blue points are in the corners
of the cube outside the sphere.
The boundary between red and blue traces the curved surface of the sphere.

```python
# Cell 06 - 3D scatter plot: red = inside sphere, blue = outside

fig = plt.figure("mc_sphere", figsize=(7, 7))
ax = fig.add_subplot(111, projection="3d")
ax.view_init(azim=-72, elev=48)
ax.scatter(x_in, y_in, z_in, color="red", marker=".", s=0.1)
ax.scatter(x_out, y_out, z_out, color="blue", marker=".", s=0.1)
ax.set_title(f"Unit Sphere - Halton ({total_dots:,} points)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.tight_layout()
plt.show()
```

**Estimating the sphere volume.**
The cube has volume $2^3 = 8$, so:

$$\hat{V} = 8 \times \frac{\text{dots inside}}{\text{dots total}}$$

The exact volume is $\frac{4}{3}\pi \approx 4.188790$.

```python
# Cell 07 - Compute the sphere volume estimate and error

act = 4 / 3 * np.pi
est = 8 * np.count_nonzero(d <= 1.0) / total_dots
err = np.abs((est - act) / act)

print(f"dots total  : {total_dots:,}")
print(f"dots inside : {np.count_nonzero(d <= 1.0):,}")
print(f"actual      : {act:.6f}  (= 4pi/3)")
print(f"estimated   : {est:.6f}")
print(f"abs error   : {err:.5%}")
```
