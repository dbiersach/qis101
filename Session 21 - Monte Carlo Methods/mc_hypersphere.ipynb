**Monte Carlo estimation of the 4D unit hypersphere content.**
The pattern from the 2D circle and 3D sphere extends naturally to any dimension.
The general formula for the volume (**content**) of an $n$-dimensional unit ball is:

$$V_n = \frac{\pi^{n/2}}{\Gamma\!\left(\frac{n}{2}+1\right)}$$

For $n = 4$, $\Gamma(3) = 2! = 2$, giving:

$$V_4 = \frac{\pi^2}{\Gamma(3)} = \frac{\pi^2}{2} \approx 4.934802$$

The bounding hypercube $[-1,1]^4$ has content $2^4 = 16$, so:

$$\frac{V_4}{16} = \frac{\pi^2/2}{16} = \frac{\pi^2}{32}
\quad\Rightarrow\quad
V_4 \approx 16 \times \frac{\text{dots inside}}{\text{dots total}}$$

A point $(x,y,z,w)$ is inside the unit 4-ball if $x^2+y^2+z^2+w^2 \leq 1$.

**There is no scatter plot in this notebook** - 4-dimensional space cannot be
directly visualized.
This is one of the key advantages of Monte Carlo: the algorithm is identical
regardless of dimension; only the distance formula and the bounding
hypercube volume change.
Note that 6,250,000 samples are used here (50× more than the 3D sphere)
because higher-dimensional spaces have proportionally more volume
near the corners of the hypercube, so more points fall outside the ball
and fewer contribute to the estimate.

**Halton sequence (Numba-compiled).**
Four dimensions require four distinct prime bases: 2, 3, 5, 7.

```python
"""mc_hypersphere.ipynb"""

# Cell 01 - Numba-compiled Halton quasi-random sequence

import numpy as np
import pandas as pd
from numba import float64, int64, vectorize


@vectorize([float64(int64, int64)], nopython=True)
def halton(n, p):
    h, f = 0, 1
    while n > 0:
        f = f / p
        h += (n % p) * f
        n = int(n / p)
    return h


print(halton(np.int64(10_000), np.int64(2)))
```

**Sample count.**
6,250,000 points - 50× more than the 3D sphere notebook.
In 4D the ball occupies $V_4/16 = \pi^2/32 \approx 30.8\%$ of the hypercube,
compared to $(4\pi/3)/8 \approx 52.4\%$ in 3D,
so more points are needed to maintain accuracy.

```python
# Cell 02 - Number of sample points

total_dots = 6_250_000
print(f"{total_dots = :,}")
```

**Generating 4D Halton sample points.**
Four coordinates use four distinct prime bases (2, 3, 5, 7)
so all dimensions are independent.
Each coordinate is scaled from $(0,1]$ to $(-1, 1]$.

```python
# Cell 03 - Generate 4D Halton sample points in [-1, 1]^4
# x: base 2, y: base 3, z: base 5, w: base 7

n_arr = np.arange(total_dots, dtype=np.int64)

x = (1 - halton(n_arr, np.full(total_dots, 2, dtype=np.int64))) * 2 - 1
y = (1 - halton(n_arr, np.full(total_dots, 3, dtype=np.int64))) * 2 - 1
z = (1 - halton(n_arr, np.full(total_dots, 5, dtype=np.int64))) * 2 - 1
w = (1 - halton(n_arr, np.full(total_dots, 7, dtype=np.int64))) * 2 - 1

pd.DataFrame({"x": x[:5], "y": y[:5], "z": z[:5], "w": w[:5]})
```

**4D distance from the origin.**
The Euclidean distance formula extends to any number of dimensions:
$d = \sqrt{x^2 + y^2 + z^2 + w^2}$.
A point is inside the unit 4-ball if and only if $d \leq 1$.

```python
# Cell 04 - 4D Euclidean distance from the origin

d = np.sqrt(x**2 + y**2 + z**2 + w**2)

pd.DataFrame({"x": x[:5], "y": y[:5], "z": z[:5], "w": w[:5], "d": d[:5]})
```

**Estimating $V_4 = \pi^2/2$.**
The hypercube $[-1,1]^4$ has content $2^4 = 16$, so:
$\hat{V}_4 = 16 \times (\text{dots inside} / \text{dots total})$.

```python
# Cell 05 - Estimate the 4D hypersphere content and compute error

act = np.pi**2 / 2
est = 16 * np.count_nonzero(d <= 1.0) / total_dots
err = np.abs((est - act) / act)

print(f"dots total  : {total_dots:,}")
print(f"dots inside : {np.count_nonzero(d <= 1.0):,}")
print(f"actual      : {act:.6f}  (= pi^2/2)")
print(f"estimated   : {est:.6f}")
print(f"abs error   : {err:.5%}")
```
