**Positional notation and base conversion.**
Every integer can be represented in any base $b$ as a weighted sum of powers:

$$
(d_{n-1}\, d_{n-2}\, \ldots\, d_1\, d_0)_b
= \sum_{k=0}^{n-1} d_k \cdot b^k
$$

where $d_0$ is the **least significant digit** (rightmost) and
$d_{n-1}$ is the **most significant digit** (leftmost).
In binary ($b=2$) the digits are restricted to $\{0, 1\}$ and each
position represents a power of two.

This notebook first builds binary-to-decimal and decimal-to-binary
converters from scratch to make the algorithm explicit, then shows
the equivalent Python built-ins that handle the same task in one call.

```python
"""base_conversion.ipynb"""

# Cell 01 - Binary to decimal: evaluate the positional weighted sum


def binary_to_decimal(b_str: str) -> int:
    """Convert a binary string to its decimal integer value.

    Iterates from the least significant bit (rightmost character) to the
    most significant bit, accumulating digit * 2**position for each bit.
    """
    total = 0
    for position, char in enumerate(reversed(b_str)):
        total += int(char) * 2**position
    return total


b_str = "1011001"
d = binary_to_decimal(b_str)
print(f"{b_str} in binary = {d:,} in decimal")
```

**Decimal to binary: repeated division by 2.**
The algorithm extracts bits from least significant to most significant
by repeatedly taking the remainder when dividing by 2 (`d % 2` gives
the current LSB) then integer-dividing by 2 to shift right.
Because bits are collected LSB-first, the accumulated string must be
reversed at the end to restore MSB-first order.

```python
# Cell 02 - Decimal to binary: extract bits via repeated division by 2


def decimal_to_binary(d: int) -> str:
    """Convert a non-negative decimal integer to its binary string representation.

    Extracts each bit as the remainder of successive integer divisions by 2.
    Bits are collected LSB-first then reversed to produce MSB-first output.
    """
    b_str = ""
    while d > 0:
        b_str += str(d % 2)  # LSB of current value
        d //= 2  # shift right by one bit
    return b_str[::-1]  # reverse: collected LSB-first, output MSB-first


d = 113
b_str = decimal_to_binary(d)
print(f"{d:,} in decimal = {b_str} in binary")
```

**Python and NumPy built-ins for base conversion.**
Python provides several concise built-in options that replace the manual
algorithms above:

- `int(s, base)` parses any string in the given base and returns a decimal integer.
- `bin(n)` returns a binary string prefixed with `'0b'`; slice `[2:]` strips the prefix.
- The f-string format spec `f"{n:b}"` produces a binary string with no prefix.
- `np.base_repr(n, base)` converts to any base (2 through 36), making it the
  most general option when bases other than 2, 8, or 16 are needed.

```python
# Cell 03 - Built-in conversions: int(), bin(), f-string, and np.base_repr()

import numpy as np

b_str = "1011001"
d = 113

# Binary string -> decimal integer
print(f"{b_str} in binary = {int(b_str, 2):,} in decimal    [int(s, 2)]")

# Decimal integer -> binary string (three equivalent built-in approaches)
print(f"{d:,} in decimal = {bin(d)[2:]} in binary   [bin(d)[2:]]")
print(f"{d:,} in decimal = {d:b} in binary   [f'{{d:b}}']")
print(
    f"{d:,} in decimal = {np.base_repr(d, base=2)} in binary   [np.base_repr(d, base=2)]"
)
```
