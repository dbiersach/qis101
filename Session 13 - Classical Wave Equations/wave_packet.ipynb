**Wave packets and the superposition principle.**
A **wave packet** is a localized disturbance formed by superimposing many
sinusoidal waves of slightly different frequencies.
In quantum mechanics, a particle's wavefunction is a wave packet: the
more waves that are combined, the more sharply localized the packet becomes
in space - at the cost of a wider spread in momentum (frequency).
This is one physical expression of the Heisenberg uncertainty principle
$\Delta x \cdot \Delta p \geq \hbar/2$.

Each component wave here is $\cos^2(f_i x)$, where the frequencies
$f_i$ are evenly spaced across $[-1, 1)$.
The running average $\frac{1}{N}\sum_{i=0}^{N-1}\cos^2(f_i x)$ is
animated frame by frame as $N$ grows, showing the packet sharpen.
The animation starts from $N=1$ and builds toward `num_total_waves`,
adding one wave per frame.

```python
"""wave_packet.ipynb"""

# Cell 01 - Wave packet animation with num_total_waves=1 (single component)

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# Enable interactive animation in Jupyter Lab
%matplotlib widget

num_total_waves = 1


def plot():
    global xa, ya, cur_wave_num, wave_data, wave_packet

    xa = np.linspace(-6 * np.pi, 6 * np.pi, 600)

    # Allocate one row per component wave (not 2x - second half was never used)
    wave_data = np.empty((num_total_waves, 600))

    # Frequencies evenly spaced across [-1, 1) - one per component
    for i in range(num_total_waves):
        f = 2 * (i / num_total_waves - 0.5)
        wave_data[i] = np.cos(f * xa) ** 2

    cur_wave_num = 1
    ya = np.sum(wave_data[:cur_wave_num], axis=0) / cur_wave_num
    (wave_packet,) = ax.plot(xa, ya, color="blue")

    ax.set_title(f"Wave Packets = {cur_wave_num}")
    ax.set_xlabel("Location")
    ax.set_ylabel("Probability (Not Normalized)")
    ax.set_ylim(-0.1, 1.1)


def anim_frame_counter():
    global cur_wave_num
    while cur_wave_num < num_total_waves:
        cur_wave_num += 1
        yield cur_wave_num


def anim_draw_frame(n):
    global ax, ya, wave_packet
    ya = np.sum(wave_data[:cur_wave_num], axis=0) / cur_wave_num
    wave_packet.set_data(xa, ya)
    ax.set_title(f"Wave Packets = {n}")
    return (wave_packet,)


def plot_waves():
    global ax, anim
    plt.close("all")
    plt.figure("wave_packet")
    ax = plt.axes()
    plot()
    anim = FuncAnimation(
        ax.figure,
        anim_draw_frame,
        anim_frame_counter,
        interval=25,
        repeat=False,
        cache_frame_data=False,
    )
    plt.show()


plot_waves()
```

**Three component waves.**
With 3 waves the packet is still very broad - the average of three
cosine-squared functions with slightly different frequencies produces
only a gentle modulation envelope.

```python
# Cell 02 - 3 component waves: packet begins to form

num_total_waves = 3
plot_waves()
```

**Eleven component waves.**
With 11 waves a recognizable packet shape emerges at the center.
Constructive interference concentrates amplitude near $x=0$ while
destructive interference suppresses the wings.

```python
# Cell 03 - 11 component waves: clear packet shape emerging

num_total_waves = 11
plot_waves()
```

**Two hundred component waves.**
With 200 waves the packet is tightly localized near $x=0$.
The broad frequency spread (many $f_i$ values covering $[-1, 1)$) produces
a narrow spatial extent - a direct illustration of the uncertainty principle:
a well-defined position requires a wide spread of momenta.

```python
# Cell 04 - 200 component waves: tightly localized packet

num_total_waves = 200
plot_waves()
```
