**The Pauli gates and the single-qubit gate set.**
A quantum gate is a **unitary** matrix $U$ satisfying $UU^\dagger = I$.
Unitarity guarantees that gate operations preserve the norm of a state vector -
total probability is always conserved.
Note that unitary is a weaker condition than Hermitian: every Hermitian unitary
satisfies $U = U^\dagger$, but the S and T gates below are unitary without being Hermitian.

The four **Pauli matrices** $\{I, X, Y, Z\}$ are both Hermitian and unitary.
They form a basis for the space of all $2 \times 2$ Hermitian matrices and appear
throughout quantum computing as the fundamental single-qubit gates.
$I$ is the identity (wire, no-op).
$X$ flips $|0\rangle \leftrightarrow |1\rangle$ and is the quantum NOT gate.
$Z$ leaves $|0\rangle$ unchanged and maps $|1\rangle \to -|1\rangle$ (phase flip).
$H$ maps $|0\rangle \to |+\rangle$ and $|1\rangle \to |-\rangle$, creating equal superpositions.
The cells below verify a set of algebraic identities that are used repeatedly
when simplifying quantum circuits.

```python
"""quantum_gates.ipynb"""

# Cell 01 - Identity #1:  X² = Y² = Z² = I

import numpy as np
from IPython.display import Math, display

from qis101_utils import as_latex

g_I = np.array([[1, 0], [0, 1]], dtype=complex)
g_X = np.array([[0, 1], [1, 0]], dtype=complex)
g_Y = np.array([[0, -1j], [1j, 0]], dtype=complex)
g_Z = np.array([[1, 0], [0, -1]], dtype=complex)

display(as_latex(g_I, prefix=r"\mathbf{I}="))
display(as_latex(g_X, prefix=r"\mathbf{X}="))
display(as_latex(g_Y, prefix=r"\mathbf{Y}="))
display(as_latex(g_Z, prefix=r"\mathbf{Z}="))

display(as_latex(g_X @ g_X, prefix=r"\mathbf{X^2}="))
display(as_latex(g_Y @ g_Y, prefix=r"\mathbf{Y^2}="))
display(as_latex(g_Z @ g_Z, prefix=r"\mathbf{Z^2}="))

display(Math(r"\mathbf{Identity\;\#1}"))
display(
    Math(
        r"\mathbf{X^2}=\mathbf{I}\;?\;\rightarrow\;"
        f"{np.isclose(g_X @ g_X, g_I).all()}"
    )
)
display(
    Math(
        r"\mathbf{Y^2}=\mathbf{I}\;?\;\rightarrow\;"
        f"{np.isclose(g_Y @ g_Y, g_I).all()}"
    )
)
display(
    Math(
        r"\mathbf{Z^2}=\mathbf{I}\;?\;\rightarrow\;"
        f"{np.isclose(g_Z @ g_Z, g_I).all()}"
    )
)
```

**The Hadamard gate.**
The Hadamard gate $H$ creates equal superpositions from basis states and is
its own inverse: $H^2 = I$.
It can be expressed as a normalized sum of the $X$ and $Z$ Pauli gates:

$$
H = \frac{1}{\sqrt{2}}(X + Z) = \frac{1}{\sqrt{2}}\begin{bmatrix}1&1\\1&-1\end{bmatrix}
$$

This relationship shows that $H$ lies exactly halfway between $X$ (bit flip)
and $Z$ (phase flip) on the Bloch sphere.

```python
# Cell 02 - Identity #2:  H = (1/√2)(X + Z)

g_H = (1 / np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=complex)

display(as_latex(g_H, prefix=r"\mathbf{H}="))
display(
    as_latex(
        1 / np.sqrt(2) * (g_X + g_Z),
        prefix=r"\mathbf{\frac{1}{\sqrt{2}}(X+Z)}=",
    )
)

display(Math(r"\mathbf{Identity\;\#2}"))
display(
    Math(
        r"\mathbf{H}=\mathbf{\frac{1}{\sqrt{2}}(X+Z)}\;?\;\rightarrow\;"
        f"{np.isclose(g_H, 1 / np.sqrt(2) * (g_X + g_Z)).all()}"
    )
)
```

**Hadamard conjugation.**
Sandwiching a Pauli gate between two Hadamards transforms it into another Pauli.
This is the quantum analogue of a change of basis: $H$ swaps the roles of the
$X$ and $Z$ axes on the Bloch sphere, so $HZH = X$, $HXH = Z$, and $HYH = -Y$.
These identities are used constantly in circuit simplification and in converting
between the $X$- and $Z$-basis measurements.

```python
# Cell 03 - Identity #3:  HZH = X

display(as_latex(g_X, prefix=r"\mathbf{X}="))
display(as_latex(g_H @ g_Z @ g_H, prefix=r"\mathbf{HZH}="))

display(Math(r"\mathbf{Identity\;\#3}"))
display(
    Math(
        r"\mathbf{X}=\mathbf{HZH}\;?\;\rightarrow\;"
        f"{np.isclose(g_X, g_H @ g_Z @ g_H).all()}"
    )
)
```

```python
# Cell 04 - Identity #4:  HXH = Z

display(as_latex(g_Z, prefix=r"\mathbf{Z}="))
display(as_latex(g_H @ g_X @ g_H, prefix=r"\mathbf{HXH}="))

display(Math(r"\mathbf{Identity\;\#4}"))
display(
    Math(
        r"\mathbf{Z}=\mathbf{HXH}\;?\;\rightarrow\;"
        f"{np.isclose(g_Z, g_H @ g_X @ g_H).all()}"
    )
)
```

```python
# Cell 05 - Identity #5:  HYH = -Y

display(as_latex(-g_Y, prefix=r"\mathbf{-Y}="))
display(as_latex(g_H @ g_Y @ g_H, prefix=r"\mathbf{HYH}="))

display(Math(r"\mathbf{Identity\;\#5}"))
display(
    Math(
        r"\mathbf{-Y}=\mathbf{HYH}\;?\;\rightarrow\;"
        f"{np.isclose(-g_Y, g_H @ g_Y @ g_H).all()}"
    )
)
```

```python
# Cell 06 - Identity #6:  XYX = -Y

display(as_latex(-g_Y, prefix=r"\mathbf{-Y}="))
display(as_latex(g_X @ g_Y @ g_X, prefix=r"\mathbf{XYX}="))

display(Math(r"\mathbf{Identity\;\#6}"))
display(
    Math(
        r"\mathbf{-Y}=\mathbf{XYX}\;?\;\rightarrow\;"
        f"{np.isclose(-g_Y, g_X @ g_Y @ g_X).all()}"
    )
)
```

**Phase gates: S and T.**
The S gate (phase gate) and T gate ($\pi/8$ gate) apply a phase rotation to
the $|1\rangle$ component only, leaving $|0\rangle$ unchanged:

$$
S = \begin{bmatrix}1&0\\0&i\end{bmatrix} \qquad
T = \begin{bmatrix}1&0\\0&e^{i\pi/4}\end{bmatrix}
$$

Both are unitary but neither is Hermitian ($S^\dagger \neq S$).
The T gate is a "half-S": applying it twice recovers the S gate ($T^2 = S$),
and applying S twice gives the Z gate ($S^2 = Z$).
Together with H, the T gate generates a gate set that is **universal** for
quantum computation - any unitary operation can be approximated to arbitrary
precision using only $\{H, T\}$.

```python
# Cell 07 - Identity #7:  T² = S

g_S = np.array([[1, 0], [0, 1j]], dtype=complex)
g_T = np.array([[1, 0], [0, np.exp(1j * np.pi / 4)]], dtype=complex)

display(as_latex(g_S, prefix=r"\mathbf{S}="))
display(as_latex(g_T, prefix=r"\mathbf{T}="))
display(as_latex(g_T @ g_T, prefix=r"\mathbf{T^2}="))

display(Math(r"\mathbf{Identity\;\#7}"))
display(
    Math(
        r"\mathbf{S}=\mathbf{T^2}\;?\;\rightarrow\;"
        f"{np.isclose(g_S, g_T @ g_T).all()}"
    )
)
```
