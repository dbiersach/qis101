**The X gate is self-inverse.**
A gate $U$ is **self-inverse** (or **involutory**) if $U^{-1} = U$,
which is equivalent to $U^2 = I$.

The Pauli-X gate is:

$$X = \begin{bmatrix}0&1\\1&0\end{bmatrix}$$

It is both **Hermitian** ($X = X^\dagger$) and **unitary** ($XX^\dagger = I$).
For any Hermitian unitary matrix these two properties together imply
self-inverse: $X^{-1} = X^\dagger = X$.

This notebook extracts the X gate matrix using Qiskit's `Operator` class,
computes its matrix inverse numerically via `numpy.linalg.inv`,
and verifies that $X^{-1} = X$.

```python
"""x_gate_inverse.ipynb"""

# Cell 01 - Extract the X gate unitary matrix via Qiskit's Operator class

import numpy as np
from IPython.display import Math, display
from qiskit.circuit.library import XGate
from qiskit.quantum_info import Operator

from qis101_utils import as_latex

x_matrix = Operator(XGate()).data  # extract the 2x2 unitary matrix

display(as_latex(x_matrix, prefix=r"\mathbf{X}="))
```

---
**Computing the matrix inverse.**
`numpy.linalg.inv` computes the general matrix inverse numerically.
For the X gate the result should equal X itself.

```python
# Cell 02 - Compute the matrix inverse of X

x_inv_matrix = np.linalg.inv(x_matrix)

display(as_latex(x_inv_matrix, prefix=r"\mathbf{X^{-1}}="))
```

---
**Verifying self-inverse.**
`np.allclose` checks element-wise equality within floating-point tolerance,
confirming $X^{-1} = X$.

```python
# Cell 03 - Verify X == X_inv (self-inverse / involutory)

is_self_inverse = np.allclose(x_matrix, x_inv_matrix)

display(
    Math(
        r"\mathbf{X}="
        r"\mathbf{X^{-1}}"
        rf"\;?\;\rightarrow\;{is_self_inverse}"
    )
)
```

---
**Determinant of the X gate.**
Every quantum gate is unitary, so its determinant is a complex number of
magnitude 1. The group of all $2\times2$ unitary matrices is called $U(2)$.

A gate belongs to the **special** unitary group $SU(2)$ only if its
determinant is exactly $+1$. Computing the determinant tells us which group
X lives in:

$$\det X = \det\begin{bmatrix}0&1\\1&0\end{bmatrix}
= (0)(0) - (1)(1) = -1$$

Because $\det X = -1$ (not $+1$), the X gate is unitary but **not** in
$SU(2)$. The same is true of all three Pauli gates.

We can always promote a $U(2)$ gate into $SU(2)$ by multiplying by a global
phase $e^{i\theta}$, since for a $2\times2$ matrix
$\det(e^{i\theta}X) = e^{i2\theta}\det X$. Choosing $\theta = \pi/2$ gives
the factor $i$, and

$$\det(i\ X) = i^{2}\det X = (-1)(-1) = +1$$

so $i\ X$ **is** in $SU(2)$. A global phase like this has no observable effect
on measurement outcomes, so $iX$ acts on qubits exactly like $X$.

```python
# Cell 04 - Determinant of X (checks U(2) vs SU(2) membership)

det_x = np.linalg.det(x_matrix)

# A gate is in SU(2) only when its determinant equals +1
is_special_unitary = np.isclose(det_x, 1.0)

display(Math(rf"\det \mathbf{{X}} = {det_x.real:+.0f}"))

# Promote X into SU(2) by applying the global phase i (theta = pi/2)
x_su2_matrix = np.asarray(x_matrix, dtype=complex) * 1j
det_x_su2 = np.linalg.det(x_su2_matrix)
is_su2 = np.isclose(det_x_su2, 1.0)

display(as_latex(x_su2_matrix, prefix=r"i\ \mathbf{X}="))
display(Math(rf"\det (i\ \mathbf{{X}}) = {det_x_su2.real:+.0f}"))
```

---
**Eigenvalues and eigenvectors of the X gate.**
An eigenvector $\lvert v\rangle$ of X satisfies $X\lvert v\rangle
= \lambda\lvert v\rangle$, where the scalar $\lambda$ is its eigenvalue

Because X is Hermitian, its eigenvalues are real, and because X is unitary,
they have magnitude 1. The only real numbers of magnitude 1 are $+1$ and
$-1$, so those must be the eigenvalues. Their eigenvectors are the familiar
$\lvert + \rangle$ and $\lvert - \rangle$ states:

$$\lvert + \rangle = \tfrac{1}{\sqrt{2}}\begin{bmatrix}1\\1\end{bmatrix}
\;(\lambda = +1), \qquad
\lvert - \rangle = \tfrac{1}{\sqrt{2}}\begin{bmatrix}1\\-1\end{bmatrix}
\;(\lambda = -1)$$

`numpy.linalg.eig` returns the eigenvalues and the matrix whose columns are
the corresponding (unit-length) eigenvectors

```python
# Cell 05 - Eigenvalues and eigenvectors of X

eigenvalues, eigenvectors = np.linalg.eig(x_matrix)

display(as_latex(eigenvalues, prefix=r"\text{eigenvalues}="))
display(as_latex(eigenvectors, prefix=r"\text{eigenvectors (as columns)}="))

# Display each eigenvalue next to its (column) eigenvector for a clear pairing
for i, value in enumerate(eigenvalues):
    vector = eigenvectors[:, i]
    display(
        as_latex(
            vector,
            column=True,
            prefix=rf"\lambda = {value.real:+.0f}\;\rightarrow\;",
        )
    )
```
