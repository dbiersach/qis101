**Real vectors and the dot product.**
A **vector** is an ordered list of numbers (its components) that represents
both a magnitude and a direction.
In NumPy a vector is a 1-D array.
This notebook introduces three key properties:

- **Norm** (magnitude): $\|\mathbf{v}\| = \sqrt{\sum_i v_i^2}$
- **Dot product**: $\mathbf{v} \cdot \mathbf{w} = \sum_i v_i w_i = \|\mathbf{v}\|\|\mathbf{w}\|\cos\theta$
- **Orthonormal basis**: a set of unit vectors that are mutually perpendicular,
  whose dot products with any vector extract that vector's components

```python
"""real_vectors.ipynb"""

# Cell 01 - Create a 2D unit vector v at angle theta1 = pi/4

import numpy as np
from IPython.display import Math, display

from qis101_utils import as_latex


def print_ndarray_info(name, a):
    print(f"Type of {name} is {type(a).__name__}")
    print(f"Number of dimensions of {name} = {a.ndim}")
    print(f"Shape of dimensions of {name} = {a.shape}")
    print(f"Length of {name} = {len(a)}")
    print(f"Size of {name} = {a.size}")


theta1 = 1 / 4 * np.pi  # 45 degrees
theta2 = 3 / 4 * np.pi  # 135 degrees

# Create a NumPy 1-D array (vector) using the angle theta1
v = np.array([np.cos(theta1), np.sin(theta1)])

print_ndarray_info("v", v)
display(as_latex(v, prefix=r"\mathbf{v}="))
```

**A second unit vector.**
`w` is constructed the same way at angle $\theta_2 = 3\pi/4$ (135 degrees).
Both `v` and `w` lie on the unit circle, so both have norm 1.

```python
# Cell 02 - Create a second 2D unit vector w at angle theta2 = 3*pi/4

w = np.array([np.cos(theta2), np.sin(theta2)])

print_ndarray_info("w", w)
display(as_latex(w, prefix=r"\mathbf{w}="))
```

**The norm (magnitude) of a vector.**
`np.linalg.norm` computes $\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2}$.
Since both vectors were defined as $[\cos\theta, \sin\theta]$,
the Pythagorean identity $\cos^2\theta + \sin^2\theta = 1$ guarantees
both norms equal exactly 1 - they are **unit vectors**.

```python
# Cell 03 - Norm (magnitude) of both vectors

display(Math(rf"\|\mathbf{{v}}\|={np.linalg.norm(v)}"))
display(Math(rf"\|\mathbf{{w}}\|={np.linalg.norm(w)}"))
```

**The dot product and floating-point roundoff.**
The angle between `v` (at 45°) and `w` (at 135°) is exactly 90°,
so the dot product $\mathbf{v}\cdot\mathbf{w} = \cos(90°) = 0$.
However, floating-point arithmetic produces a tiny nonzero residual
($\approx -10^{-17}$) rather than exact zero.
`np.round` is used to suppress the roundoff for display,
and `np.arccos` of the normalized dot product recovers the angle.

```python
# Cell 04 - Dot product, roundoff demonstration, and angle between vectors

v_dot_w = np.dot(v, w)

# Raw result shows floating-point roundoff (shown in red)
display(Math(rf"\mathbf{{v\cdot w}}=\color{{red}}{{{v_dot_w}}}"))
# Rounded to 5 decimal places suppresses the noise
display(Math(rf"\mathbf{{v\cdot w}}={np.round(v_dot_w, 5)}"))

theta = np.degrees(np.arccos(np.dot(v, w) / (np.linalg.norm(v) * np.linalg.norm(w))))
display(Math(rf"\theta={theta}^\circ"))
```

**The standard Cartesian basis.**
The three unit basis vectors $\hat{\mathbf{i}}$, $\hat{\mathbf{j}}$, $\hat{\mathbf{k}}$
are **orthonormal**: each has norm 1 and all pairs are perpendicular
($\hat{\mathbf{i}}\cdot\hat{\mathbf{j}} = 0$, etc.).
Any vector in $\mathbb{R}^3$ can be written as a linear combination of these three.

```python
# Cell 05 - The standard orthonormal Cartesian basis vectors in R^3

i = np.array([1, 0, 0])
j = np.array([0, 1, 0])
k = np.array([0, 0, 1])

display(as_latex(i, prefix=r"\mathbf{\hat{i}}="))
display(as_latex(j, prefix=r"\mathbf{\hat{j}}="))
display(as_latex(k, prefix=r"\mathbf{\hat{k}}="))
```

**Extracting components via dot products.**
Taking the dot product of any vector with a basis vector extracts
the component along that basis direction.
This works because $\mathbf{v}\cdot\hat{\mathbf{i}} = v_x\cdot 1 + v_y\cdot 0 + v_z\cdot 0 = v_x$.
This is the geometric meaning of the dot product: **projection** onto a unit vector.

```python
# Cell 06 - Extract vector components by projecting onto each basis vector

v = np.array([2, 3, 5])

display(as_latex(v, prefix=r"\mathbf{v}="))
display(Math(rf"\mathbf{{v}}\cdot\mathbf{{\hat{{i}}}}={np.dot(v, i)}"))
display(Math(rf"\mathbf{{v}}\cdot\mathbf{{\hat{{j}}}}={np.dot(v, j)}"))
display(Math(rf"\mathbf{{v}}\cdot\mathbf{{\hat{{k}}}}={np.dot(v, k)}"))
```
