**Dirac notation and quantum state vectors.**
In quantum mechanics every physical state is represented by a *ket* $|\psi\rangle$ —
a complex-valued column vector whose entries are called **probability amplitudes**.
The probability of measuring outcome $k$ is

$$
P(k) = \frac{|\psi_k|^2}{\|\psi\|^2}
$$

where $|\psi_k|$ is the modulus of the $k$-th amplitude and $\|\psi\|$ is the
Euclidean norm of the full state vector.
Note that the *phase* of an amplitude (its angle in the complex plane) carries no
probability information on its own — only the modulus squared matters.

```python
"""dirac_notation.ipynb"""

# Cell 01 - Complex amplitudes and measurement probability

import numpy as np
from IPython.display import Math, display

from qis101_utils import as_latex

psi = np.array([-3 - 1j, -2j, 1j, 2])
norm_psi = np.linalg.norm(psi)

prob_2 = abs(psi[2]) ** 2 / norm_psi**2

display(as_latex(psi, prefix=r"\mathbf{\lvert\psi\rangle}=", column=True))
display(Math(rf"\textrm{{Probability of outcome 2 = {prob_2 * 100:.4f}\%}}"))
```

**Global phase invariance.**
Multiplying a ket by any nonzero complex scalar $s$ produces a physically
equivalent state: the scalar cancels in the ratio $|\psi_k|^2 / \|\psi\|^2$,
so every measurement probability is unchanged.
This is the quantum-mechanical principle of **global phase invariance** —
the states $|\psi\rangle$ and $s|\psi\rangle$ are physically identical.

```python
# Cell 02 - Multiplying a ket by a scalar leaves probabilities unchanged

s = -2.3 - 11j
s_psi = s * psi

prob_2_scaled = abs(s_psi[2]) ** 2 / np.linalg.norm(s_psi) ** 2

display(Math(rf"\mathbf{{s=}}\;{s}"))
display(as_latex(s_psi, prefix=r"\mathbf{s\lvert\psi\rangle}=", column=True))
display(Math(rf"\textrm{{Probability of outcome 2 = {prob_2_scaled * 100:.4f}\%}}"))
```

**Normalization.**
A ket is *normalized* when $\|\psi\| = 1$, which simplifies the probability
formula to just $P(k) = |\psi_k|^2$.
Dividing every amplitude by the norm produces a normalized ket without changing
any probabilities.
For a normalized ket the squared amplitudes must sum to 1:

$$
\sum_k \left|\frac{\psi_k}{\|\psi\|}\right|^2 = 1
$$

```python
# Cell 03 - Normalizing a ket

psi = np.array([2 - 3j, 1 + 2j])
norm_psi = np.linalg.norm(psi)
normed_psi = psi / norm_psi

# Squared moduli of the normalized amplitudes — must sum to 1.
# Take .real to suppress the negligible imaginary residual from floating-point arithmetic.
probs = (normed_psi * normed_psi.conj()).real

display(as_latex(psi, prefix=r"\mathbf{\lvert\psi\rangle}=", column=True))
display(Math(rf"\mathbf{{\|\psi\|}}=\;{norm_psi:.4f}"))
display(
    as_latex(
        normed_psi,
        prefix=r"\mathbf{\frac{\lvert\psi\rangle}{\|\psi\|}}=",
        column=True,
    )
)
display(
    as_latex(
        np.round(probs, 4),
        prefix=r"\mathbf{P(k)}=",
        column=True,
    )
)
```

**Transition amplitude and the inner product.**
The *bra* $\langle\phi|$ is the conjugate transpose of the ket $|\phi\rangle$.
The **bra-ket** (inner product) $\langle\phi|\psi\rangle$ measures the overlap
between two states.
Dividing by the product of their norms gives the normalized **transition amplitude** $\mathcal{A}$:

$$
\mathcal{A} = \frac{\langle\phi|\psi\rangle}{\|\phi\|\;\|\psi\|}
$$

The **transition probability** — the likelihood of a system prepared in state
$|\psi\rangle$ being found in state $|\phi\rangle$ — is then $|\mathcal{A}|^2$.
When both kets are already normalized this simplifies to $|\langle\phi|\psi\rangle|^2$.

The result $P(\psi\to\phi) = 1$ below is not a coincidence.
Inspect the two vectors: $|\phi\rangle = -i\,|\psi\rangle$, so they differ by
nothing more than the global phase factor $e^{i\theta} = -i$.
This connects directly back to Cell 02 — two kets related by a global phase
are the **same physical state**, and a transition probability of 1 is the
mathematical confirmation of that identity.

```python
# Cell 04 - Transition amplitude between two states

# Start state
psi = np.array([1, -1j])

# End state
phi = np.array([1j, 1])

bra_phi = phi.conj()
bracket_phi_psi = np.dot(bra_phi, psi)

norm_psi = np.linalg.norm(psi)
norm_phi = np.linalg.norm(phi)
amp = bracket_phi_psi / (norm_phi * norm_psi)
trans_prob = abs(amp) ** 2

display(
    as_latex(
        psi, prefix=r"\text{Start state:}\;\mathbf{\lvert\psi\rangle}=", column=True
    )
)
display(
    as_latex(phi, prefix=r"\text{End state:}\;\mathbf{\lvert\phi\rangle}=", column=True)
)
display(Math(rf"\langle\phi\lvert\psi\rangle=\;{bracket_phi_psi:.4f}"))
display(
    Math(
        rf"\mathcal{{A}}=\frac{{\langle\phi\lvert\psi\rangle}}"
        rf"{{\|\phi\|\;\|\psi\|}}=\;{amp:.4f}"
    )
)
display(Math(rf"P(\psi\to\phi)=|\mathcal{{A}}|^2=\;{trans_prob:.4f}"))
```
