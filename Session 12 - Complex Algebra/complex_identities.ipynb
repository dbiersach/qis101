**Complex number identities.**
This notebook verifies six fundamental identities of complex arithmetic
numerically.
Each identity is confirmed with `np.isclose` rather than `==` because
floating-point operations accumulate rounding errors that make exact
bit equality unreliable - the `==` comparison is shown alongside in red
to illustrate when it fails.

```python
"""complex_identities.ipynb"""

# Cell 01 - The modulus respects multiplication: |c1||c2| = |c1*c2|

import numpy as np
from IPython.display import Math, display

c1 = -2.6 + 0.7j
c2 = 5.11 - 4.9j

display(Math(f"c_1 = {c1}"))
display(Math(f"c_2 = {c2}"))
display(Math(f"|c_1| = {abs(c1)}"))
display(Math(f"|c_2| = {abs(c2)}"))

t1 = abs(c1) * abs(c2)
t2 = abs(c1 * c2)

display(Math(f"|c_1||c_2| = {t1}"))
display(Math(f"|c_1 c_2| = {t2}"))

# == tests exact bit equality - can fail due to floating-point rounding
display(Math(rf"|c_1||c_2| = |c_1 c_2|\;?\;\rightarrow\;\color{{red}}{{{t1 == t2}}}"))
# np.isclose tolerates small rounding differences
display(Math(rf"|c_1||c_2| \approx |c_1 c_2|\;?\;\rightarrow\;{np.isclose(t1, t2)}"))
```

**The triangle inequality.**
In the complex plane, adding two complex numbers is equivalent to placing
two vectors tip-to-tail.
The length of the resultant vector can never exceed the sum of the
individual lengths:

$$|c_1 + c_2| \leq |c_1| + |c_2|$$

*Equality* holds only when both numbers point in the same direction
(same argument).

```python
# Cell 02 - The triangle inequality: |c1+c2| <= |c1| + |c2|

t1 = abs(c1 + c2)  # modulus of the sum
t2 = abs(c1) + abs(c2)  # sum of the moduli

display(Math(f"|c_1 + c_2| = {np.round(t1, 5)}"))
display(Math(f"|c_1| + |c_2| = {np.round(t2, 5)}"))
display(Math(rf"|c_1+c_2| \leq |c_1|+|c_2|\;?\;\rightarrow\;{t1 <= t2}"))
```

**Conjugation respects addition.**
The conjugate of a sum equals the sum of the conjugates:

$$\overline{c_1 + c_2} = \overline{c_1} + \overline{c_2}$$

This follows directly from the definition: negating the imaginary part
distributes over addition.

```python
# Cell 03 - Conjugation respects addition

t1 = c1.conjugate() + c2.conjugate()
t2 = (c1 + c2).conjugate()

display(Math(rf"\overline{{c_1}} + \overline{{c_2}} = {np.round(t1, 5)}"))
display(Math(rf"\overline{{c_1 + c_2}} = {np.round(t2, 5)}"))
display(
    Math(
        rf"\overline{{c_1}} + \overline{{c_2}} = \overline{{c_1+c_2}}"
        rf"\;?\;\rightarrow\;{np.isclose(t1, t2)}"
    )
)
```

**Conjugation respects multiplication.**
The conjugate of a product equals the product of the conjugates:

$$\overline{c_1 \cdot c_2} = \overline{c_1} \cdot \overline{c_2}$$

```python
# Cell 04 - Conjugation respects multiplication

t1 = c1.conjugate() * c2.conjugate()
t2 = (c1 * c2).conjugate()

display(Math(rf"\overline{{c_1}} \cdot \overline{{c_2}} = {np.round(t1, 5)}"))
display(Math(rf"\overline{{c_1 \cdot c_2}} = {np.round(t2, 5)}"))
display(
    Math(
        rf"\overline{{c_1}} \cdot \overline{{c_2}} = \overline{{c_1 \cdot c_2}}"
        rf"\;?\;\rightarrow\;{np.isclose(t1, t2)}"
    )
)
```

**Complex multiplication as rotation.**
Multiplying by $e^{i\theta}$ rotates a complex number by angle $\theta$
in the complex plane.
Adding angles in the exponent is equivalent to composing two rotations:

$$e^{i(\theta_1+\theta_2)} = e^{i\theta_1}\cdot e^{i\theta_2}$$

Here $\theta_1 + \theta_2 = \frac{3\pi}{8} + \frac{\pi}{8} = \frac{\pi}{2}$,
so the product should equal $i$ (a $90°$ rotation).
The small real part in the output is floating-point noise.

```python
# Cell 05 - Complex multiplication is angular rotation

theta1 = np.radians(67.5)  # 3*pi/8
theta2 = np.radians(22.5)  # pi/8

display(Math(rf"\theta_1 = {np.round(theta1, 8)}"))
display(Math(rf"\theta_2 = {np.round(theta2, 8)}"))

t1 = np.exp(1j * (theta1 + theta2))
t2 = np.exp(1j * theta1) * np.exp(1j * theta2)

# Small real part is floating-point noise; true value is 0+1j
display(Math(rf"e^{{i(\theta_1+\theta_2)}} = {t1}"))
display(Math(rf"(e^{{i\theta_1}})(e^{{i\theta_2}}) = {t2}"))

display(
    Math(
        rf"e^{{i(\theta_1+\theta_2)}} = (e^{{i\theta_1}})(e^{{i\theta_2}})"
        rf"\;?\;\rightarrow\;\color{{red}}{{{t1 == t2}}}"
    )
)
display(
    Math(
        rf"e^{{i(\theta_1+\theta_2)}} \approx (e^{{i\theta_1}})(e^{{i\theta_2}})"
        rf"\;?\;\rightarrow\;{np.isclose(t1, t2)}"
    )
)
```

**Roots of unity.**
The $n$-th roots of unity are the $n$ complex numbers $z$ satisfying $z^n = 1$.
They are equally spaced around the unit circle at angles
$\theta_k = \frac{2\pi k}{n}$ for $k = 0, 1, \ldots, n-1$:

$$z_k = e^{2\pi i k/n} = \cos\frac{2\pi k}{n} + i\sin\frac{2\pi k}{n}$$

`np.power(c, 1/n)` returns only the principal root ($k=0$).
The general formula below recovers all $n$ roots.

```python
# Cell 06 - The five 5th roots of unity

c = 1 + 0j
n = 5

display(Math(f"c = {c}"))
display(
    Math(rf"\text{{Principal root: }} c^{{1/{n}}} = \sqrt[{n}]{{{c}}} = {c ** (1 / n)}")
)
display(Math(rf"\text{{All }} {n} \text{{ roots}}"))

for k in range(n):
    root = abs(c) ** (1 / n) * np.exp(1j * (np.angle(c) + 2 * np.pi * k) / n)
    display(
        Math(
            rf"k={k}: z_k = {np.round(root, 5)}"
            rf"\quad\Rightarrow\quad z_k^{{{n}}} = {np.round(root**n, 5)}"
        )
    )
```
