**Phase encoding: reaching the six cardinal states of the Bloch sphere.**
The Bloch sphere has six **cardinal states** - the $\pm$ eigenstates
of the three Pauli operators:

| State | Vector | Bloch point |
|---|---|---|
| $\lvert 0\rangle$ | $[1,\;0]^T$ | North pole ($+Z$) |
| $\lvert 1\rangle$ | $[0,\;1]^T$ | South pole ($-Z$) |
| $\lvert +\rangle$ | $[1,\;1]^T/\sqrt{2}$ | $+X$ |
| $\lvert -\rangle$ | $[1,\;-1]^T/\sqrt{2}$ | $-X$ |
| $\lvert i\rangle$ | $[1,\;i]^T/\sqrt{2}$ | $+Y$ |
| $\lvert{-i}\rangle$ | $[1,\;-i]^T/\sqrt{2}$ | $-Y$ |

This notebook demonstrates that **phase-encoded states** - vectors with a
non-trivial global or relative phase such as $[-1,0]^T$, $[i,0]^T$, etc. -
sit at the same Bloch sphere point as the corresponding standard state.
The Bloch sphere visualization makes the global phase physically irrelevant:
two vectors that differ only by $e^{i\phi}$ map to the same point.

---
Each circuit uses `save_statevector` at intermediate steps so we can
watch the state move across the Bloch sphere as each gate is applied.
A single `AerSimulator` backend is shared across all cells.

```python
"""phase_encoding.ipynb"""

# Cell 01 - Imports and shared AerSimulator backend

import numpy as np
from IPython.display import display
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_bloch_multivector
from qiskit_aer import AerSimulator

from qis101_utils import as_latex

backend = AerSimulator()
```

---
**Cell 02 - Reaching $[-1,\;0]^T$ from $|0\rangle$.**
The sequence $X \to Z \to X$ applied to $|0\rangle$:
- $X|0\rangle = |1\rangle = [0,1]^T$ (south pole)
- $ZX|0\rangle = Z|1\rangle = -|1\rangle = [0,-1]^T$ (south pole, global phase)
- $XZX|0\rangle = X(-|1\rangle) = -|0\rangle = [-1,0]^T$ (north pole, global phase)

The final state $[-1,0]^T = e^{i\pi}|0\rangle$ is the north pole with a global phase
of $\pi$ - physically identical to $|0\rangle$.

```python
# Cell 02 - Create the [-1, 0] state via X -> Z -> X

qc = QuantumCircuit(1)
qc.x(0)
qc.save_statevector(label="sv1")
qc.z(0)
qc.save_statevector(label="sv2")
qc.x(0)
qc.save_statevector(label="sv3")

display(qc.draw(output="mpl"))

result = backend.run(transpile(qc, backend)).result()
sv1 = result.data(0)["sv1"]
sv2 = result.data(0)["sv2"]
sv3 = result.data(0)["sv3"]

display(as_latex(sv1, prefix=r"\mathbf{Statevector\;1}="))
display(plot_bloch_multivector(sv1))
display(as_latex(sv2, prefix=r"\mathbf{Statevector\;2}="))
display(plot_bloch_multivector(sv2))
display(as_latex(sv3, prefix=r"\mathbf{Statevector\;3}="))
display(plot_bloch_multivector(sv3))
```

---
**Cell 03 - Reaching $[i,\;0]^T$ from $|0\rangle$.**
The sequence $R_x(\pi/2) \to R_y(\pi/2) \to R_z(-\pi/2) \to R_y(\pi/2)$
navigates the Bloch sphere to produce $[i,0]^T = e^{i\pi/2}|0\rangle$,
which is the north pole with a global phase of $\pi/2$.

```python
# Cell 03 - Create [i, 0] state via Rx(pi/2) -> Ry(pi/2) -> Rz(-pi/2) -> Ry(pi/2)

qc = QuantumCircuit(1)
qc.rx(np.pi / 2, 0)
qc.save_statevector("sv1")
qc.ry(np.pi / 2, 0)
qc.save_statevector("sv2")
qc.rz(-np.pi / 2, 0)
qc.save_statevector("sv3")
qc.ry(np.pi / 2, 0)
qc.save_statevector("sv4")

display(qc.draw(output="mpl"))

result = backend.run(transpile(qc, backend)).result()
sv1 = result.data(0)["sv1"]
sv2 = result.data(0)["sv2"]
sv3 = result.data(0)["sv3"]
sv4 = result.data(0)["sv4"]

display(as_latex(sv1, prefix=r"\mathbf{Statevector\;1}="))
display(plot_bloch_multivector(sv1))
display(as_latex(sv2, prefix=r"\mathbf{Statevector\;2}="))
display(plot_bloch_multivector(sv2))
display(as_latex(sv3, prefix=r"\mathbf{Statevector\;3}="))
display(plot_bloch_multivector(sv3))
display(as_latex(sv4, prefix=r"\mathbf{Statevector\;4}="))
display(plot_bloch_multivector(sv4))
```

---
**Cell 04 - Reaching $[-i,\;0]^T$ from $|0\rangle$.**
The sequence $R_x(-\pi/2) \to R_y(\pi/2) \to R_z(\pi/2) \to R_y(\pi/2)$
produces $[-i,0]^T = e^{-i\pi/2}|0\rangle$ - again the north pole, global phase $-\pi/2$.

```python
# Cell 04 - Create [-i, 0] state via Rx(-pi/2) -> Ry(pi/2) -> Rz(pi/2) -> Ry(pi/2)

qc = QuantumCircuit(1)
qc.rx(-np.pi / 2, 0)
qc.save_statevector("sv1")
qc.ry(np.pi / 2, 0)
qc.save_statevector("sv2")
qc.rz(np.pi / 2, 0)
qc.save_statevector("sv3")
qc.ry(np.pi / 2, 0)
qc.save_statevector("sv4")

display(qc.draw(output="mpl"))

result = backend.run(transpile(qc, backend)).result()
sv1 = result.data(0)["sv1"]
sv2 = result.data(0)["sv2"]
sv3 = result.data(0)["sv3"]
sv4 = result.data(0)["sv4"]

display(as_latex(sv1, prefix=r"\mathbf{Statevector\;1}="))
display(plot_bloch_multivector(sv1))
display(as_latex(sv2, prefix=r"\mathbf{Statevector\;2}="))
display(plot_bloch_multivector(sv2))
display(as_latex(sv3, prefix=r"\mathbf{Statevector\;3}="))
display(plot_bloch_multivector(sv3))
display(as_latex(sv4, prefix=r"\mathbf{Statevector\;4}="))
display(plot_bloch_multivector(sv4))
```

**Cell 05 - Every quantum circuit before measurement is a single unitary matrix.**
Because unitary matrices are closed under multiplication
($U_1 U_2$ is unitary whenever $U_1$ and $U_2$ are unitary),
any sequence of gates applied to a qubit can always be collapsed into
one composite unitary $\hat{O}$.
For a circuit with gates $G_1, G_2, \ldots, G_n$ applied left to right,
the composite unitary is the **right-to-left matrix product**:

$$\hat{O} = G_n \cdots G_2 \cdot G_1$$

(Right-to-left because matrix-vector multiplication applies the rightmost
matrix first: $\hat{O}|\psi\rangle = G_n(\cdots G_2(G_1|\psi\rangle)\cdots)$.)

Qiskit's `save_unitary()` instruction asks the `AerSimulator` to compute
this product automatically during simulation.
`result.get_unitary(qc)` then retrieves the resulting $2\times 2$ matrix.
Applying $\hat{O}$ to $|0\rangle$ manually confirms it produces the same
final state as the step-by-step simulation in Cell 04.

```python
# Cell 05 - Composite unitary of the [-i, 0] circuit
#
# The four gates Rx(-pi/2), Ry(pi/2), Rz(pi/2), Ry(pi/2) each have a
# 2x2 unitary matrix. Their product (right-to-left) is the single unitary
# O = Ry(pi/2) @ Rz(pi/2) @ Ry(pi/2) @ Rx(-pi/2)
# that describes the entire circuit in one step.
#
# save_unitary() tells AerSimulator to compute this product during simulation.
# result.get_unitary(qc) retrieves the resulting 2x2 matrix.
# Multiplying O @ |0> then reproduces the [-i, 0] final state.

qc = QuantumCircuit(1)
qc.rx(-np.pi / 2, 0)
qc.ry(np.pi / 2, 0)
qc.rz(np.pi / 2, 0)
qc.ry(np.pi / 2, 0)
qc.save_unitary()

display(qc.draw(output="mpl"))

result = backend.run(transpile(qc, backend)).result()
op = result.get_unitary(qc)

initial_state = np.array([[1], [0]])
final_state = op @ initial_state

display(as_latex(op, prefix=r"\large\mathbf{\hat{O}}="))
display(as_latex(final_state, prefix=r"\large\mathbf{\hat{O}\lvert 0\rangle}="))
```

**Cell 06 - Reaching $[0,\;-1]^T$ from $|0\rangle$.**
The sequence $X \to Z$ applied to $|0\rangle$:
- $X|0\rangle = |1\rangle = [0,1]^T$
- $ZX|0\rangle = -|1\rangle = [0,-1]^T$

The final state $[0,-1]^T = e^{i\pi}|1\rangle$ is the south pole with global phase $\pi$,

```python
# Cell 06 - Create [0, -1] state via X -> Z

qc = QuantumCircuit(1)
qc.x(0)
qc.save_statevector("sv1")
qc.z(0)
qc.save_statevector("sv2")

display(qc.draw(output="mpl"))

result = backend.run(transpile(qc, backend)).result()
sv1 = result.data(0)["sv1"]
sv2 = result.data(0)["sv2"]

display(as_latex(sv1, prefix=r"\mathbf{Statevector\;1}="))
display(plot_bloch_multivector(sv1))
display(as_latex(sv2, prefix=r"\mathbf{Statevector\;2}="))
display(plot_bloch_multivector(sv2))
```

**Cell 09 - Composite unitary for the $[0,\;-i]^T$ circuit.**
Same approach as Cell 05: `save_unitary()` collapses the four-gate sequence
into one $2\times 2$ matrix, which is then applied to $|0\rangle$ to confirm
the $[0,-i]^T$ result.

```python
# Cell 09 - Composite unitary of the [0, -i] circuit
# See Cell 05 for the full explanation of how save_unitary() and
# get_unitary() collapse a gate sequence into a single 2x2 matrix.

qc = QuantumCircuit(1)
qc.rx(np.pi / 2, 0)
qc.ry(np.pi / 2, 0)
qc.rz(-np.pi / 2, 0)
qc.ry(-np.pi / 2, 0)
qc.save_unitary()

display(qc.draw(output="mpl"))

result = backend.run(transpile(qc, backend)).result()
op = result.get_unitary(qc)

initial_state = np.array([[1], [0]])
final_state = op @ initial_state

display(as_latex(op, prefix=r"\large\mathbf{\hat{O}}="))
display(as_latex(final_state, prefix=r"\large\mathbf{\hat{O}\lvert 0\rangle}="))
```
