```python
"""qiskit_packages.ipynb"""

# Cell 01 - Verify the Qiskit SDK is importable and print its version

import qiskit

print(f"Qiskit SDK Version: {qiskit.__version__}")
```

**Listing all installed Qiskit-related packages.**
`importlib.metadata.distributions()` iterates over every package installed
in the active environment.
Filtering by `'qiskit'` in the package name and sorting alphabetically
gives a complete inventory of the Qiskit ecosystem components
(core SDK, Aer simulator, IBM Runtime, machine learning, optimization, etc.).

```python
# Cell 02 - List all installed Qiskit-related packages and their versions

from importlib.metadata import distributions

qiskit_packages = sorted(
    [pkg for pkg in distributions() if "qiskit" in pkg.metadata["Name"].lower()],
    key=lambda pkg: pkg.metadata["Name"].lower(),
)

for pkg in qiskit_packages:
    print(f"Package: {pkg.metadata['Name']:<30} Version: {pkg.version}")
```
