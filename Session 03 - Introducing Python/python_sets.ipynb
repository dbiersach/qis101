```python
"""python_sets.ipynb"""

# Cell 01 - Create & Read

a = {10, 20, 30, 40}
b = {30, 40, 50, 60}
c = {10, 20, 20, 10, 30}  # notice duplicates

print(f"{a=}")  # Python controls the order of set items
print(f"{b=}")  # Output may not match the input order!
print(f"{c=}  # duplicates removed")

print(f"len(a) = {len(a)}")
try:
    first_item = a[0]  # This will fail because sets are unordered
except TypeError as e:
    print("Error:", e)
print("Is 20 in a?", 20 in a)
print("Is 99 in a?", 99 in a)
```

```python
# Cell 02 - Update, Delete, Remove. Discard, Pop

try:
    # Sets don't support indexing, so we can't access items by position
    a[0] = 7  # This will fail because sets are unordered
except TypeError as e:
    print("Error:", e)
a.add(50)
a.add(20)  # duplicate; has no effect
print(f"{a=}")  # 20 was already in the set
a.remove(30)  # removes 30 (error if missing)
print(f"{a=}")
a.discard(999)  # safe, no error if missing
print(f"{a=}")
popped_item = a.pop()
print(f"Popped item: {popped_item}")
print(f"{a=}")
```

```python
# Cell 03 - Clearing Sets

print(f"{b=}")
b.clear()
print(f"{b=}")  # Set is now empty
b = set()  # Declare a NEW empty set
print(f"{b=}")
```

```python
# Cell 04 - Sets support heterogeneous data types

mixed = {"Alice", 42, 3.14, True}
print(f"{mixed=}")
```

```python
# Cell 05 - Looping through a Set

d = {100, 200, 300}
for item in d:
    print("Set item:", item)
```

```python
# Cell 06 - Union, Intersection, Difference

x = {1, 2, 3, 4}
y = {3, 4, 5, 6}
print(f"{x=}")
print(f"{y=}")
print("Union (x | y):", x | y)
print("Intersection (x & y):", x & y)
print("Difference (x - y):", x - y)
print("Difference (y - x):", y - x)
print("Symmetric Difference (x ^ y):", x ^ y)
```

```python
# Cell 07 - Demonstrate Duplicate Removal

duplicates = {1, 1, 2, 2, 3, 3, 3}
print(f"{duplicates=}\n{len(duplicates)=}")
```
