```python
"""python_tuples.ipynb"""

# Cell 01 - C = Create and R = Read

a = [10, 20, 30]  # Variable a is a List
b = (50, 60, 70, 80)  # Variable b is a Tuple

print(f"{a=}")
print(f"{b=}")
print(f"{len(a)=}, {len(b)=}")
print(f"{a[0]=}, {a[-1]=}, {a[1:]=}")
print(f"{b[0]=}, {b[-1]=}, {b[1:]=}")
```

```python
# Cell 02 - U = Update, D = Delete

a[1] = 99
a.append(40)
print(f"{a=}")
try:
    # This will cause an error
    b[1] = 99  # Tuples are immutable
except TypeError as e:
    print("Error:", e)
print(f"{b=}")
try:
    # This will cause an error
    del b[1]  # Tuples are immutable
except TypeError as e:
    print("Error:", e)
# "delete" first element by creating a NEW tuple
b = b[1:]
print(f"{b=}")
```

```python
# Cell 03 - Insert and Append

try:
    # This will cause an error
    b.insert(2, 30)  # Tuples are immutable
except AttributeError as e:
    print("Error:", e)
try:
    # This will cause an error
    b.append(50)  # Tuples are immutable
except AttributeError as e:
    print("Error:", e)
b = b + (60,)  # "append" by creating a NEW tuple
print(f"{b=}")  # tuples can have duplicate items
```

```python
# Cell 04 - Clearing Lists and Tuples

print(f"{a=}")
a.clear()
print(f"{a=}")
try:
    b.clear()  # Tuples are immutable
except AttributeError as e:
    print("Error:", e)
a = []  # A NEW empty list
b = ()  # A NEW empty tuple
print(f"{a=}")
print(f"{b=}")
```

```python
# Cell 05 - Tuples support heterogeneous data types

b = ("Alice", 42, 3.14, True)
print(f"{b=}")
```

```python
# Cell 06 - Looping through a Tuple

for i in range(len(b)):
    print(i, b[i])
print()
for index, item in enumerate(b):
    print(index, item)
```
