```python
"""python_dictionaries.ipynb"""


from pprint import pprint

# Cell 01 - Create & Read

capitals = {
    "USA": "Washington D.C.",
    "Egypt": "Cairo",
    "India": "New Delhi",
    "China": "Beijing",
    "Russia": "Moscow",
    "France": "Paris",
    "Morocco": "Rabat",
    "Germany": "Berlin",
    "South Africa": "Pretoria",
}
print(capitals)  # Regular print may not be well-formatted
pprint(capitals)  # Pretty Print for better readability
print(capitals["USA"])  # Accessing a value by key
```

```python
# Cell 02 - U = Update, D = Delete

capitals["USA"] = "New York"
del capitals["Russia"]
pprint(capitals)
```

```python
# Cell 03 - Insert

capitals.update({"United Kingdom": "London"})
capitals["Spain"] = "Madrid"
pprint(capitals)
```

```python
# Cell 04 - Iterating through a Dictionary

for country in capitals:  # Looping via keys
    print(f"Country: {country}, Capital: {capitals[country]}")
print()

for country, capital in capitals.items():  # Looping via items
    print(f"Country: {country}, Capital: {capital}")
```

```python
# Cell 05 - Dictionary Components

keys = capitals.keys()
values = capitals.values()
print("Countries:", list(keys))
print("Capitals:", list(values))
```
