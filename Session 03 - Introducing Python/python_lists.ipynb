```python
"""python_lists.ipynb"""

# Cell 01 - C = Create and R = Read

a = [10, 20, 30, 40, 50]
print(f"{a=}")
print(f"{a[0]=}, {a[4]=}, {a[-1]=}, {a[-2]=}")
```

```python
# Cell 02 - U = Update, D = Delete

a[0] = 15
print(f"{a=}")
del a[2]
print(f"{a=}")
```

```python
# Cell 03 - Insert and Append

a.insert(2, 30)
print(f"{a=}")
a.append(60)
a.append(20)  # Lists can have duplicate items
print(f"{a=}")
```

```python
# Cell 04 - Extending Lists

a.extend([70, 80, 90])
print(f"{a=}")
```

```python
# Cell 05 - Slicing Lists

print(f"{a[1:4]=}")
print(f"{a[:2]=}")
print(f"{a[2:]=}")
print(f"{a[::2]=}")
print(f"{a[1:5:2]=}")
print(f"{a[1::2]=}")
```

```python
# Cell 06 - Lists with Mixed Data Types

b = ["Alice", 42, 3.14, True]
print(f"{b=}")
```

```python
# Cell 07 - Looping via index number

for i in range(len(b)):
    print(i, b[i])
```

```python
# Cell 08 - Looping via enumerate

for index, item in enumerate(b):
    print(index, item)
```
